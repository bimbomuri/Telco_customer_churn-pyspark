```python
from pyspark.ml.linalg import Vectors
from pyspark.ml.feature import VectorAssembler
```

```python

assembler=VectorAssembler(inputCols=['account length','area code','number vmail messages','total day minutes',
                                     'total day calls','total day charge','total eve minutes','total eve calls','total night minutes',
                                     'total night calls','total night charge','total intl minutes',
                                     'total intl calls','total intl charge','customer service calls'],outputCol='features')
 
output_data=assembler.transform(df_data)
```

```python
output_data.printSchema()
```

```python
final_data=output_data.select('features','churn')
```

```python
from pyspark.ml.classification import LogisticRegression
```

```python
train,test=final_data.randomSplit([0.7,0.3])
```

```python
model=LogisticRegression(maxIter=10, regParam=0.3, elasticNetParam=0.8,labelCol='churn')
```

```python
model=model.fit(train)
```

```python
summary=model.summary
summary.predictions.describe().show() 
```

```python
from pyspark.ml.evaluation import MulticlassClassificationEvaluator
predictions=model.evaluate(test)
evaluator=MulticlassClassificationEvaluator(metricName="accuracy",predictionCol='prediction',labelCol='churn')
 
accuracy=evaluator.evaluate(predictions.predictions)
 
model1=LogisticRegression(maxIter=10, regParam=0.3, elasticNetParam=0.8,labelCol='churn')
model1=model1.fit(final_data)
 
test_data=assembler.transform(df_data)
print("Accuracy = %g " % ( accuracy))
```
